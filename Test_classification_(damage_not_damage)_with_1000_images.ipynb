```python
import cv2
import os
import shutil
from keras.models import load_model
import numpy as np
import matplotlib.pyplot as plt
from keras.preprocessing import image
```

```python
path = 'C:/Users/pc/Desktop/test img/'
```

```python
# model = load_model("C:/Users/pc/Desktop/jupyter/imageclassifier_v1.h5")
```

```python
model = load_model("C:/Users/pc/Desktop/new_90%_vgg16_classifier.h5")
```

```python
damage_list = []
whole_list = []
for i in os.listdir(path):
    for k in os.listdir(path + i):
        if i == 'damage':
            damage_list.append(k)
        else :
            whole_list.append(k)
```

```python
# len(whole_list)
len(damage_list)
damage_list
```

```python
true_damage_counter = 0
false_damage_counter = 0
true_whole_counter = 0
false_whole_counter = 0

for sub_dir in os.listdir(path):
    for file in os.listdir(path + sub_dir):
        img = image.load_img(path + sub_dir + '/' + file, target_size=(224,224))
    #     plt.imshow(img)
    #     plt.show()
        X = image.img_to_array(img)
        val = model.predict(np.expand_dims(X, 0))

        if np.argmax(val) == 0 :
            if sub_dir == 'damage' and file in damage_list:
                true_damage_counter += 1
            else :
                false_damage_counter += 1
            shutil.move(path + sub_dir + '/' + file, 'C:/Users/pc/Desktop/test result/damage/' + file)
            print(file + ' removed to test result/damage')
        else :
            if sub_dir == 'whole' and file in whole_list:
                true_whole_counter += 1
            else :
                false_whole_counter += 1
            shutil.move(path + sub_dir + '/' + file, 'C:/Users/pc/Desktop/test result/whole/' + file)
            print(file + ' removed to test result/whole')
```

```python
total_images = true_damage_counter + false_damage_counter + true_whole_counter + false_whole_counter
print('results of classification damage/not damage with 1000 images')
print('True damage images :' + str(true_damage_counter))
print('False damage images :' + str(false_damage_counter))
print('True whole images :' + str(true_whole_counter))
print('False whole images :' + str(false_whole_counter))
print('Total images : ' + str(total_images))
```
